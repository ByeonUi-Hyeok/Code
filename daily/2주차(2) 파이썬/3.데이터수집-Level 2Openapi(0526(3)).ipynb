```python
# 응답결과를 딕셔너리 객체로 만든다
# json > 딕셔너리로 만들어 내는 모듈이 있을 것이다.
#아래가 그 모듈 결과 부분에 있던거
import json
```

```python
# 뉴스를 네이버에 요청해서 결과를 가져올 수 있는 URL이 필요하다.
# 네이버 검색 Open API 예제 긁어온거(도큐먼트>서비스aip>검색)
import os
import sys
import urllib.request
#import 뒤에 라이브러리(모듈)임
#모듈 = *.py
#패키지 = 디렉터리
#모듈 + 패키지 = 라이브러리(모듈)
#import는 쓰기 직전 아무데나 써도 된다.
#------------------------------------------------------------------------------------------------------------
# client_id = "YOUR_CLIENT_ID"
# client_secret = "YOUR_CLIENT_SECRET"
# 위의 기존코드는 우리가 위에 만든 클라이언트아이디 시크릿 만들었기에 삭제
#-------------------------------------------------------------------------------------------------------------------
#뉴스 검색어
# searchKeyword는 검색어를 저장할 변수 
# 검색어는 한글,영,숫자,특문 등 가능
# abcABC가나다123!@#
# 한글을 서버로 전송하면 기본적으로 한글이 깨진다 > 정확한 정보전달을 위해 인코딩 처리가 필요하다.
searchKeyword = '한미정상회담'
#------------------------------------------------------------------------------------------------
encText = urllib.parse.quote(searchKeyword)
# urllib.parse.quote는 인코딩 처리하는 함수이다.
# 검색할단어를 인코딩해서 enText에 담는다.
# 인코딩하면 > 특수한문자%ED등 으로 바꾸고 서버는 코드를 받아서 한글로 디코딩한다.
# 만약 urlib.parse.quote(한미정상회담123@!)를 하면
# %ED%95%9C%EB%AF%B8%EC%A0%95%EC%83%81%ED%9A%8C%EB%8B%B412aB%21%40
# 이렇게  출력
# 숫자 알파벳은 그대로 출력
#----------------------------------------------------------------------------------------------
# url에 대놓고 보이면서 데이터를 검색하는 방식이 get방식이다.

url = 'https://openapi.naver.com/v1/search/news.json'
#json 포멧 뉴스 끌어오기 주소
# document > 서비스api > 검색> 뉴스에 해당내용 있음. (dev.naver.com)
#---------------------------------------------------------------------------------------------
#데이터(파라미터)를 서버에 전달하는 방법? Method
#Method의 방식 = get, post, put, delete
#형식 : URL?키= 값  // 파이썬의 딕셔너리같다
#여기서 값 = 데이터,겅색어, 파라미터
#위의 뉴스에 3.요청변수보면 쿼리는 필수여부 y 되어있음 쿼리는 꼭 들어가야함
# URL?query=값(데이터,검색어,파라미터)
# https://openapi.naver.com/v1/search/news.json?query= 한미정상회담의 특수한 문자
# url = url + '?query=' enText
# 문자열 포멧팅 %, format(), f ''
# 상황, 취향에맞게 3개중에 골라쓰면된다.
url = f'{url}?query={encText}'
 #https://openapi.naver.com/v1/search/news.json?query=%ED%95%9C%EB%AF%B8%EC%A0%95%EC%83%81%ED%9A%8C%EB%8B%B4
 #의 결과가 나오는데 저걸로 들어가면 인증 실패나온다
 #  위의 시크릿키를 안넣었기 때문
 # ---------------------------------------------------------------------------------
 # 파이썬의 모든 것은 객체다?
 # - -----------------------------------------------------
 # 요청 객체 생성한다 클래스는 첫글자가 대문자
# urlib 패키지
#request 패키지
# request 클래스
request = urllib.request.Request(url)
# 헤더에 인증키를 추가한다.(최대한 숨겨서 서버로 전송)
request.add_header("X-Naver-Client-Id",CLIENT_ID)
request.add_header("X-Naver-Client-Secret",CLIENT_SECRET)
#-----------------------------------------------------

# 요청 - > 인터넷 - > 서버 -> 내부과정 -> 응답
# 요청하는코드 (URLOPEN은 주소창에 주소치고 엔터치는것과같음)
response = urllib.request.urlopen(request)
#--------------------------------------------------===========
#응답 결과확인
#현재 제일 최신 기사들이나옴
rescode = response.getcode()
if(rescode==200):
  #아래 라이브러리를 돌리면 결과가 딕셔너리로 나오게된다.
  tmp = json.load(response)
  # 응답결과를 딕셔너리 객체로 만든다
  # json > 딕셔너리로 만들어 내는 모듈이 있을 것이다.
    # 헤드라인만 뽑으려면? 결과 나온것  json형식이기때문에 
#----------------------------------------------------------------
    # response_body = response.read()
    # print(response_body.decode('utf-8'))
#---------------------------------------- 네이버에서 준 기존코드 삭제
else:
    print("Error Code:" + rescode)

print(type(tmp), end ='\n')
print(tmp)
#이렇게하면  json이 dic형식으로 바뀌고 10개의 키를 가지고 나온다
# 여기서 해드라인만 뽑아쓰려면??
# http://json.parser.online.fr/
# 위사이트 홀따음표말고 쌍따음표로 된 json결과를 편하게 볼수있게 정리해줌
#------------------------------------------------------------------------------
#대화형 툴은 한줄 여러번 실행해도 상관 없다.

# 헤드라인만 뽑기
# 딕셔너리의 인덱싱
# 변수 [ 키 ]
type(tmp['items']), len(tmp['items'])
print(type(tmp[ 'items' ]), len(tmp['items'])
#결과는 리스트이다.
#리스트는 포문과 많이 쓴다.
# tmp[items]의 타입과 길이를 구한다.
# 리스트, 10개
#길이 제한있을때 for , 무한 while
for item in tmp['items']:
  #뒤의 tmp아이템즈에서 앞의 임시변수 item에 하나씩 옮긴다 안의 내용량 만큼
  print(type(item))
  #item변수의 타입은 무엇인가
  print(item['title'])
  # item 딕셔너리의 키 title의 밸류값을 뽑는다
  #헤드라인은 title로 되있다.
```
