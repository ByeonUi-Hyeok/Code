# 데이터 분석 절차

- 통상적으로 5 ~ 6단계로
정립되어 있다.

## 연구 목표 수립
 

 - 전제조건 (해당 주제에 대한 데이터가 존재 하는가? (어떤 주제이든))
 - 시스템 통합(산출물)을 고려하여서 목표를 수립

## 데이터 수집

 - 4차 산업의 원유, 프로젝트의 전제조건
    - 강화학습은 예외(알파고)
 - 데이터 수집 난이도에 따라 4단계로 분류
 - 만약 데이터가 매일/매시간/매분 **지속적으로 수집해야하는 데이터라면(주기성을 가진 데이터 / 특정 인터벌을 가진) 자동화 기술이 필요하다.**
    - OS(운영체제) 수준에서 해결
       - Windows : 작업스케줄러
       - Linux : Cron
 - 수집한 데이터는 저작권 문제 有
    - 비상업적데이터 문제 無
    - 상업적데이터 문제 有

## 데이터 준비 / 전처리 / 정제

 - 데이터 준비 > 분석 > 모델 의 과정은 ***계속해서 반복된다.(에자일 개발방식)***
 - ***데이터의 품질을 향상(적합하게 구성) 시키는데 목적이 있음***
 - 분석(통계 등)
    - 관계없다
    - 그래도 ***수치화를 지향*** 해야함.

 - 머신러닝
    - 관계없다
    - 단, 학습의 효과 / 효율이 높게 나오는 방법은 ***데이터를 수치화*** 하는 것

 - 딥러닝
    - 모든 데이터는 ***수치화*** 되어야 함. (수치화 시키기위해 전략을 구성함)
    - Why >  딥러닝의 모든 학습 행위의 기본은 행/열 연산 > ***행/열***은 수치로만 계산된다.

 - 작업
    - 결측치 제거
    - 이상치 처리
    - 라벨링
    - 독립변수(Feature / 특성)
    - 종속변수(Lable / 클래스 / 정답)
    - 파생변수
    - 기타 등등
 - 대표적 모듈(라이브러리)
    - NUMPY(넘파이, 넘피)
       - 수학 / 과학용 라이브러리, 푸리에, 선형대수학
    - PANDAS(팬더스, 판다스)
       - 얘의 기반은 넘파이
       - 데이터 분석의 핵심 라이브러리
          - R언어에 대응해서 만든 모듈( 팬더스 + 머신러닝 )
          - R < - > Python( 팬더스 + 머신러닝모듈)
             - R과 파이썬대응하기위해서는 파이썬의 팬더스 라이브러리와 머신러닝 모듈을 사용한다

## 모델 구축(통계 모델 / 머신러닝 or 딥러닝 모델)

 - 통계 모델
    - 통계적 기법으로 만든 모델
 - 예측 모델 (머신러닝 / 딥러닝 활용)
    - 머린러닝 / 딥러닝으로 만든 모델
    - 머신러닝의 학습법
       - 지도학습
          - 데이터가 있는데 정답도 있다.(답이있는 데이터)
          - 분류, 회귀
          - 분류 : 남/녀 딱딱 떨어지는거
          - 회귀 : 기름값이 1500즘 될거같은데? 얼마정도? 이런게 회귀
       - 비지도학습
          - 데이터가 있는데 답이 없다.(답이 없는 데이터)
          - 군집화(클러스터리링), 차원 축소
          - 넷플릭스에 들어갔는데 나는 무슨고객인가? 답이없다 이것이 비지도학습
       - 준지도 학습
          - 답이 일부 있다
          - 데이터가 있는데 일부는 답이있고 답이없다.
          - 지도 + 비지도
       - 강화 학습
          - 알파고의 학습 방법
       - 딥러닝
          - 인공지능 프로그래밍을 수행한다.
          - CNN, RNN, GAN 등
          - 엔진
             - 텐서플로우
             - 케라스
             - 파이토치
             - 위 세개가 파이썬으로 작성됨
