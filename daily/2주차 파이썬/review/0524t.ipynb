16진수 연습문제 아스키코드

```python
def show_hex_to_ch(hexa_string):
    s = hexa_string.split(' ')
#     print(s)
    for a in s:
#         print(a)
        a = int(a, 16) # 16진수를 10진수로
#         print(a)
        print(chr(a), end='')


hexa_string1 = '48 45 4C 4C 4F'
hexa_string2 = '47 47 4F 52 45 42'
    
show_hex_to_ch(hexa_string1)
show_hex_to_ch(hexa_string2)
```
