## 인덱싱 (색인, indexing)

 - 차원축소 (demension이 줄어든다)
 - 따라서 1차 배열을 인덱싱 -> 스칼라
 - 2차 배열을 인덱싱 -> 1차 배열
 - 2차 배열 -> 인덱싱 *2 -> 스칼라 (0차원??)
    - 위의것은 XXX 인덱싱 모두 공통사항
 - Syntax(문법)
    - 변수 [ 인덱스 ]              / 변수 대괄호 안에 인덱스
    - 인덱스를 사용한다 -> 순서가 존재한다. (인덱스를 쓴다는 것은 순서가 있다는 뜻)
    - 인덱스 의 종류
       - 양의 방향 ( 순방향 )
          - 0, 1, 2 ...
       - 음의 방향 ( 역방향 )
          - -1, -2, -3 ...
          



- 참고
   - 인덱스는 ~것을 특정한다는 것. 
   - 딕셔너리는 키로 ~를 특정할 수 있기때문에 인덱스 가능
   - 그러나 딕셔너리는 순서가 없기때문에 슬라이싱 불가
   - 즉, 딕셔너리는 순서는 없지만 데이터를 특정하는 용도로 키를 사용한다.
      - 키를 통해서 인덱싱 가능

```python
# 무조건 먼저 shape, dtype을 봐야한다
# shape은 얘가 어떻게 생겻나볼 수있고
# dtype은 얘의 형태를 볼 수 있음

arr, arr.shape, arr.dtype
```

```python
# 인덱싱 사용 (차원축소)
# 2D > 1D 
# ndim은 몇차원인지보는거
arr[ 0 ], arr[ 0 ].ndim
```

```python
# arr 구성원 중 값 5를 출력하라
# arr (2차원) -> 인덱싱 *2 = > 5 (0차원)

print(arr, arr[1], arr [1][1])
# 찾아가는 과정 > arr> arr의 1번인덱스> 의 1번인덱스
print(arr, arr[-1], arr [-1][-2])
# 역방향
print(arr[1][1], arr[1][-2], arr[-1][-2])
```

## 불리언 인덱싱

 - 배열의 원소들을 특정값과 비교 (조건식 활용)
    - 결과는 늘 Ture  / False 이다.
 - 조건식을 이용하여 데이터를 추출하는 방식
 - Syntax
    - 변수[ 조건식 ] 

```python
arrs = np.random.randn( 4, 10 ) 
#데이터(row)는 4개 그 데이터의 각각의 특성(colum)은 10개 로, 컬럼
arrs
```

```python
print(arrs > 0)
# 배열에 모든 원소들에 접근하여 0과 비교(조건식) = > (조건식의 결과)는 T/F이다. 
# 그렇기에 배열은 bool 배열로 나온다
```

```python
#불리언 인덱싱  인덱싱은 차원 축소 그렇기 때문에 2차원 > 1차원으로 축소
print (arrs [ arrs > 0])
# 블리언 인덱싱의 결과가 차원을 유지할려면, 추가 조치가 필요
# 필요시 진행
```

```python
# 조건식이 여러개라면 => 조건을() 묶어서 표현한다.
#  | => or => 하나만 참이면 참이다.
# syntax
# condi = (조건1) | (조건2)
condi = ( arrs > 1 ) | (arrs < -0.5)
condi
```

```python
# 조건의 해당되는 특정항목을 특정값으로 세팅 가능
arrs [ arrs < 0 ] = 0        #배열의 원소 중에 값이 0보다 작은 원소들을 모두 0으로 치환
arrs
```

## 펜시 인덱싱

## 배열의 크기가 크다면 > 압축


```python
# a,b같은 이름은 일종의 key이다. -> 딕셔너리
# 별칭
np.savez( 'arr_arch', a=arr, b=x )         # 인자들이 다 배열이다.. arr도 위에 배여ㅑㄹ, x도 위에 배열
# 배열을 원하는 만큼 다양하게 저장 가능
# 키 = 배열, 키2=배열2 등 계속 원하는만큼 넣을 수 있음.
```

```python
# arr_arch.npz
!ls arr*
```

```python
# 압축한것 로드
# with ~ as 별칭 : <= 모든 I/O에서 사용되고, 자동 close() 하는 표현

 
with np.load('arr_arch.npz') as arch :                             # arch: 별칭
  print( arch['a'] )                                                 # 별칭[키] = > 추출한다 / 키는 위에 정의한것 np.savez( 'arr_arch', a=arr, b=x ) 
  print( arch['b'] )
```

# 선형대수 - 대표적 모델?

- 곱하기 확인

```python
x = np.array( [
  [1,2,3],
  [4,5,6]              
])
y = np.array( [
  [6,11],
  [-1, 9],
  [8,2]
])
```

```python
x, x.shape
```

```python
y, y.shape
```

```python
# 내적곱 (dot product)
np.dot (x, y)

# (2,2) = > (2,3)(3,2)
# 28 = 1 * 6 + 2 * (-1) + 3 * 8 = 6 - 2 +24 = 28
```

```python
x.dot(y)
```

```python
y.dot(x)
# 전혀 다른결과가 나온다 순서가 바뀌면 
# 전혀 다른 결과, 순서가 바뀌면 결과도 다르다
```

```python
# 행렬곱(matrix product)
np.matmul (x, y)
# x.dot(y)와 결과가 같다. 우연이다. 원래다르다
```

```python
# 에러 -> broadcast  적용 않됨
# 둘의 모양이 달라서 일단 불가
x * y
```

    - 딥러닝의 기반이 되는 퍼셉트론의 계산식에 사용됨
       - y = xW + b

- 랜덤, 임의수
- 원칙 : 모든 난수를 발생하는 함수(엔진, 모듈 등등,..,)는 seed(재료)가 있어야 한다
- seed를 임의로 부여하지 않으면, 현재 시간을 시드로 사용한다(시간은 계속해서 흘러가므로, 동일한 난수가 발생되지 않는다)
- 시드를 고정하면 항상 같은 순서로 난수가 발생된다. => 재현성
- 실험환경, 학습환경등등 통제할 수 있다.
- 통상적으로 데이터를 섞을때 학습할때 등등 난수의 시드를 고정해서 사용한다.
- **통상적으로, 데이터를 섞을때, 학습할때 등등 난수의 시드를 고정해서 사용한다**
