```python
# 1. 모듈 가져오기(import)
from bs4 import BeautifulSoup
from urllib.request import  urlopen
```

```python
# 2. 타겟 사이트 지정
target_site =  'https://finance.naver.com/marketindex/exchangeList.nhn'
target_site
```

```python
# 사이트에 들어가면 서버가 html,css,js등을 내려주고 브라우저는 그걸 해석(파싱)해서 화면에 띄워줌
# 3. 접속(요청)하여 html, css, js를 다운받는다.
res = urlopen( target_site ) #이 부분은 브라우저를 띄우고 주소 치고엔터쳐서 화면이 보이기 직전과 동일하다.
```

```python
# 4. bs4(beautifulsoup)를 이용하여 html(컨텐츠를 들고있는 언어)을 파싱한다(해석한다, 메모리에 띄운다.-> DOM )
''' 
- DOM = Document Object Model
- Beautifulsoup()는 클래스이다 근거는 첫글자 대문자
- 위의 res는 html로 구성된 문자열
- 파싱을 하려면 파서(파싱(해석)하는 규칙을 가진 모듈)가 필요하다.
'''

soup = BeautifulSoup( res, 'html5lib' )          
 # html5lib : 느리더라도 확실하게 파싱하는 파서
#soup
# 결과 <html ....>
```
