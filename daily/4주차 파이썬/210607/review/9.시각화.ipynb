## 기본 그리기

```python
import numpy as np
```

- 1. 데이터 준비
- 2. 차트의 크기 준비 (모양을 잡는다) : 생략가능
- 3. 플로팅 (꾸민다) : 필수
- 4. 화면에 보여라 (주피터 노트북에 출력) : 생략가능


```python
# 1. 데이터 준비 : x, y => 2-Demension
x = np.arange( 0, 12, 0.01 ) # => 1200개의 x좌표 생성
y = np.sin( x )
# y는 x값을 sin()함수에 넣으면 구할수 있다 => 설명가능, 예측가능
# 어떤 시점이던 x를 구하기만 하면 해당식(sin)에 넣어서 y를 구할수 있다
# 이런 것이 일종이 단순 회귀 문제 이다
x, y
```

```python
(x*0).shape, (x*0)
```

```python
# 2. 차트의 크기 준비 (모양을 잡는다) : 생략가능
plt.figure(  figsize=( 12, 6 ) )

# 3. 플로팅 (꾸민다) : 필수
# 선그리기
# 2번을 그리니까, 자동으로 선의 색상을 변경해 주었다
# 한개의 차트에 여러개를 그릴경우 데이터가 구분이 않된다
# 데이터에 표기를 해야한다(범례)
plt.plot( x, y, label='sin' )
plt.plot( x, np.cos( x ), label='cos'  )

# 범례를 화면에 출력하고 싶은 경우 -> 레전드
# 개별데이터에 라벨링, 레전드 호출
plt.legend()

# x축은 무엇인가? 무슨 데이터인가? 축 라벨링
plt.xlabel('time')
plt.ylabel('value')

# 제목
plt.title(' time per value wave ')

# 배경에 격자를 줘서 값을 가늠할수 잇게 가이드 해주자
plt.grid()

# 임계값 표시 -> 수직선, 수평선, 대각선등등 사용
# 수평선을 그리시오, y는 무조건 0이다
# '--' 는 선의 모양이고, 데이터끼리 구분하거나, 경계, 기준치를 표현
# 할때 주로 많이 사용
plt.plot( x, x*0, '--', label='limit'  )

# 4. 화면에 보여라 (주피터 노트북에 출력) : 생략가능
# 여러개를 그린후, 한번에 송출할때 유용
plt.show()
```

```python
plt.figure(  figsize=( 12, 6 ) )

plt.plot( x, y, label='sin' )
plt.plot( x, np.cos( x ), label='cos'  )
plt.legend()
plt.xlabel('time')
plt.ylabel('value')
plt.title(' time per value wave ')
plt.grid()
plt.plot( x, x*0, '--', label='limit'  )

# 특정 부분만 확대해서 표현한다면
plt.xlim(4,6)
plt.ylim(-1,-0.5)

plt.show()
```

## 박스플롯

- 데이터의 분포를 보고 사각형의 박스로 표현
- 박스 내부에는 데이터의 중간값이 표시(수평선)
- 데이터가 높은값, 작은값에 일부 존재하고 대부분 중심부에서 움직인다면 위 아래는 얇은 실선으로 내려가고, 중심부를 중심으로 박스가 형성된다
- **박스가 있는 지점이 데이터가 가장 많이 밀집한 곳**이다

```python
# 차트의 모양
plt.figure( figsize=(8,6) )

# 플로팅
sns.boxplot( x=tips.total_bill)

# 보여라
plt.show()
# 해석
# 이 레스토랑(?)은 13~24$ 사이에 매출이 존재한다
# 매출의 50% 지점은 17.795000$ 이다
# 간혹, 완전싼 음식(3$)이나 고가의 음식(대량주문)을 한 고객이 
# 간간히 있다, 고가 매출에는 40불은 기점으로 소수는 그 이상을 시키고
# 40불 이하는 그래도 어느 정도 손님이 있다
```

```python
plt.figure( figsize=(8,6) )

# 컬럼명으로 x,y,hue를 지정하면, data 항목을 설정해야 한다
sns.boxplot( x='day', y='total_bill', data=tips)

# 보여라
plt.show()

# 해석
# 4개 요일 중에 일요일이 50분위값이 가장 높다(평균 매출도 높다)
# 최고 매출을 기록한 고객은 토요일에 존재
# 금요일은 매출이 가장 작다, 50분위도 작고, 간단하게 먹는 고객들이
# 많았던 것으로 예측
# 이 레스토랑은 주 4일 근무(오픈)하는 회사이다
# 단, 손님의 수(?), 흡연 여부, 성별, 팁 등의 관계는 않보인다
```

```python
plt.figure( figsize=(8,6) )
# 흡연이 매출에 미치는 영향
sns.boxplot( x='day', y='total_bill', hue='smoker', data=tips)
plt.show()
# 평일은 흡연자가 매출이 적고, 주말엔 흡연자가 더 많은 매출을 올린다
# 레스토랑은 평일과 주말에 흡연 가능 테이블의 수를 조정해야
# 최대 매출을 것이다 -> 권장(가능하다면, 단정할수는 없다)
# 50분위 위치, 데이터 분포량등 순으로 분석하여 결론에 도달
```

```python
plt.figure( figsize=(8,6) )
# 인원(size) 대비 매출
sns.boxplot( x='day', y='total_bill', hue='size', data=tips)
plt.show()
# 일요일엔 홀로족이 없다
# 인원이 많을수록 매출이 높다
# 금요일엔 2명 손님이 앞도적이다(데이트?)
# 목요일엔 6명 손님이 등장, 4명 손님도 비교적 많다
# 목요일엔 단체석을 늘리고, 금요일엔 2인석을 늘려라 (매출이 증대될듯)
```

## 산포도, 산점도

- 차트에 **데이터의 분포**를 그린다
- x와 y간의 (독립변수와 종속변수간의) 상관성, 방정식을 구현, 찾는다, 표현한다
  - x:독립변수
    - 특징(feature), 컬럼
  - y:종속변수
    - **정답**, 클레스, 레이블

```python
# 손님당 매출액(total_bill)과 팁(tip) 간에 어떤 관계가 있는가?
# 둘사이를 정의할수 있는 방정식이 있는가? -> 단순회귀의 문제
sns.lmplot( x='total_bill', y='tip', data=tips, size=7 )
plt.show()
# 해석
# 대각선의 실선은 손님당 매출과 팁에 대한 선형 단순 회귀를 
# 계산하여(방정식을 구해서, 1차함수를 구해서) 그린것이다
# 이 회귀모델에 의하면, 손님당 매출 20불이면 팁 3불을 받을 수 있다
# 이 이상 매출 혹은 이하 매출일대는 오차가 존재한다
# 오차가 발생한 이유 
# 1. 데이터의 분포가 퍼져있다(표준편차가 올라가 버림)
# 2. 데이터의 빈도가 적다(소액, 고액 손님이 적어서 부정확해저 버림)
# 방정식
# tip = a * total_bill + b  : y = ax + b
# 이런식이 구성되게 a(기울기), b(y절편)을 자동으로 구해서 시각화
```

```python
sns.lmplot( x='total_bill', y='tip', hue='smoker', data=tips, size=7 )
plt.show()
# 매출 17불 이상이면서 급연하는 고객이 팁을 더준다
```

## 산포행렬

- 변수(독립변수, 피처, 특성, 컬럼)들간의 관계를 표현한다
- 이떤 예측을 위해서 변수들이 얼마나 영향을 미치는지도 가늠된다

```python
# 데이터 로드
# 아이리스(붓꽃) 데이터 : 머신러닝>지도학습>분류 때 자주 활용됨
# 아이리스의 꽃잎 과 꽃받침의 너비, 높이 값을 추출해서( 2*2 )
# 총 4개의 특성을 이용하여서, 아이리스의 종류를 구분한 데이터
# 아이리스의 종류는 총 3개(범주형)
# 꽃잎 과 꽃받침의 너비, 높이 => 독립변수(특성,피처)
# 어이리스 종류 => 종속변수(정답, 레이블, 클레스)
# 컬럼들 중에서 어떤것은 독립변수, 어떤것은 종속변수이다
iris = sns.load_dataset('iris')
iris.head(2)
# 아래 데이터는 컬럼들중에 피처와 정답이 존재하므로
# 머신러닝 관점에서는 지도학습>분류로 해결 가능하다=>예측모델을 구성시
```

```python
iris.shape
```

```python
iris.info()
```

```python
# 정답의 종류 확인
iris.species.unique()
# 총 3개의 종류를 가지고 있다
# 'setosa', 'versicolor', 'virginica'
```

```python
# 개별적인 특성(피처, 너비/높이)들이 종을 구분하는데 얼마나 유용한지확인
# 수치 데이터만 영향을 받음
sns.pairplot( iris )
plt.show()
# 종이 구분 않됨
```

```python
sns.pairplot( iris, hue='species' )
plt.show()
# 'setosa', 'versicolor', 'virginica'
# setosa는  어떤 조합이던 잘 분류해 낸다
# 'versicolor', 'virginica' 2개의 종자는 컬럼별 조합의 가중치를
# 서로 다르게 부여하여서 해당 데이터가 어떤 종을 나타내는지 
# 학습해야 하고, 이 내용을 기반으로 차후에 예측 모델의 지표가된다
# 이렇게 될것이라고 분석하였다
# 종이 잘 구분이 않되는 조합은 종을 판단하는데 가중치를 낮추고
# 종이 잘 구분되는 컬럼의 조합은 종을 판단하는데 가중치를 높인다

# 종이 잘 구분이 않되는 조합은 종을 판단하는데
# 가중치를 낮추는게 좋다
# 종이 잘 구분되는 컬럼의 조합은 종을 판단하는데
# 가중치를 높이는게 좋다
```

```python
from string import ascii_lowercase
import os

p     = re.compile('[^a-z]*')
STD_A = ord('a')

# 종속변수를 담는 그릇
labels   = list()
# 독립변수를 담는 그릇
features = list()

for file in files:

  # 종속변수(정답, 레이블, 클레스) 추출
  label = os.path.basename( file )[:2]

  # 독립변수(피처,특성) 추출
  with open(file) as f:
    raw_data = f.read().lower()
    tmp      = p.sub('',raw_data)      
    counts   = [0]*len(ascii_lowercase)      
    for ch in tmp:
      counts[ ord(ch) - STD_A ] += 1
    count_norms = list( map( lambda x:x/len(tmp) , counts ) )
  
  # 추출된 데이터를 담기
  labels.append(   label )
  features.append( count_norms )
  
  #break
```
