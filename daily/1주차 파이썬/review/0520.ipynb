### 카드번호 맞히기

```python
# 1. 번호 입력하기
start = input('시작번호') # start = int(input('시작번호'))
end = input('마지막번호')
start = int(start)
end = int(end)

# 2. 시작 ~ 마지막 번호로 랜덤 수 추출
import random
num = random.randint(start, end)

# 3. 숫자 맞히기
while True:
    user = int(input('숫자를 입력해주세요.'))
    if user == num:
        print('정답입니다.')
        break
    else:
        if user > num:
            print('더 적은 수를 입력해주세요.')
        else:
            print('더 큰 수를 입력해주세요.')
```

```python
total = 0
count = 0
while count < len(score):
    total += score[count] # 0 1 2 3 4
    count += 1
print('합계', total, '점')
```

중복 요소는 추가하지 않는 코드

```python
# 중복 요소 발생 시 5개만 들어갈수도 있음
import random
lotto = []
for i in range(6):
    n = random.randint(1, 45)
    if n not in lotto:
        lotto.append(n)
print(lotto)
```

```python
# 절대 중복요소가 생기지 않는 코드로 변경
lotto = []
while True:
    if len(lotto) == 6:
        break
    n = random.randint(1, 45)
    if n not in lotto:
        lotto.append(n)
print(lotto)
```

야구게임

```python
import requests as req
res = req.get('http://ggoreb.com/quiz/baseball.txt')
text = res.text
print(text)
```

```python
user = '659'
strike = 0
ball = 0
# 첫번째 숫자 확인
if user[0] == text[0]:
    strike += 1 # 스트라이크
elif user[0] in text:
    ball += 1 # 볼
# 두번째 숫자 확인
if user[1] == text[1]:
    strike += 1 # 스트라이크
elif user[1] in text:
    ball += 1 # 볼
# 세번째 숫자 확인
if user[2] == text[2]:
    strike += 1 # 스트라이크
elif user[2] in text:
    ball += 1 # 볼
if strike == 0 and ball == 0:
    print('아웃')
else:
    print('%sS %sB' % (strike, ball))
```
